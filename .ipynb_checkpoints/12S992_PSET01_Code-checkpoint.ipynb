## Q1 The 100-year Storm
### 1A) What are the oods that you get exactly one 100-year storm in 100 years?

```python
def binomialdist(n,p,x):
    n_choose_x = (np.math.factorial(n)) / (np.math.factorial(x) * np.math.factorial(n-x))
    return n_choose_x * p**x * (1-p)**(n-x)
```

```python
p = 0.01 # Probability of a 100-year storm
n = 100  # Number of trials (years)
x = 1    # Number of successes

prob = binomialdist(n,p,x)
print("The probability of 1 100-year storm in 100 years is %f" % (prob))
```

### 1B) How many 100-year storms do you expect in 100 years?


```python
x         = np.arange(0,101) # Possible outcomes
probs_all = [binomialdist(n,p,num) for num in x]

print("Probability of a given number of 100-yr storms:")
print(probs_all)
print("\n")

# Evaluate the Expectation
expectation_x = np.sum(x * probs_all)
print("Expecting %f 100-yr storms in 100 years." % expectation_x)
```

### 1C) Probability of at least 1 storm in 30 years

```python
x         = np.arange(1,31,1) # Desired outcomes
n         = 30 # Set # of trials/years
probs_all = [binomialdist(n,p,num) for num in x]
prob_30   = np.sum(probs_all)

print("Probability of a given # of 100-yr storms in 30 years:")
print(probs_all)
print("\n")

print("Probability of at least 1 storm in 30-years is %f" % prob_30)
```

## Q2 90s Rock

### 2A. Mean

```python
x           = np.arange(1,13,1)
mu_x        = np.array([n*(1/12) for n in x]).sum()
print("The expected value for a d12 is %f" % mu_x)
```

### 2A. Variance

```python
var_x      = np.array([ (n - mu_x)**2 * (1/12) for n in x]).sum()
print("The variance for a d12 is %f" % var_x)
```

### 2B. Monte Carlo Simulation

```python
# Test simulation (1 iteration) and function
def throw_d12(N):
    roll_values = np.arange(1,13,1)
    rolls       = np.random.choice(roll_values,N)
    mu          = rolls.mean()
    var         = rolls.var()
    return rolls,mu,var


n_rolls    = [50,500]

# Run Simulations
sim_output = [throw_d12(n) for n in n_rolls]
[print("For %i rolls, mean is %f, variance is %f" % (n_rolls[i],sim_output[i][1],sim_output[i][2])) for i in range(2)]
```

```python
# Run Monte Carlo Simulation for each simulations

mciter     = 10000
means      = [] # [simulation][iteration]
variances  = [] # [simulation][iteration]

# Loop for simulation (50 or 500 throws)
for ii in range(2):
    
    n         = n_rolls[ii]
    means_sim = []
    vars_sim  = []
    # Loop for MC Iteration
    for mc in range(mciter):
        sim_output = throw_d12(n)
        means_sim.append(sim_output[1])
        vars_sim.append(sim_output[2])
    
    means.append(means_sim)
    variances.append(vars_sim)
```

```python
# Visualize The Histogram for the outputs
fig,axs = plt.subplots(2,2,constrained_layout=True,figsize=(8,8))
for mode in range(2):

    if mode == 0:
        plotvar  = means
        modename = "Mean"
        bins     = 25#np.arange(2,10.25,.25)
        
        plotref  = mu_x
    elif mode == 1:
        plotvar = variances
        modename = "Variance"
        bins     = 25#np.arange(5,26,1)
        plotref  = var_x
    
    for sim in range(2):
        
        ax = axs[mode,sim]
        ax.hist(plotvar[sim],bins=bins,alpha=0.55,edgecolor="w")
        ax.set_title("%ss of %i %i-roll simulations" % (modename,mciter,n_rolls[sim]))
        
        ax.axvline(plotref,color="k",label="Actual %s = %.2f" % (modename,plotref))
        ax.legend()
    #mu_sim   = np.array(plotvar[0]).mean()
    #var_sim  = np.array(plotvar[1]).mean()
    #ax.axvline(mu_sim,color='blue',label="Mean (sim.) = %.2f" % (mu_sim))
    #ax.axvline(mu_sim+2*var_sim,color='blue',ls='dashed',label="Var (sim.) = %.2f" % (var_sim))
    #ax.axvline(mu_sim-2*var_sim,color='blue',ls='dashed',label="Var (sim.) = %.2f" % (var_sim))
```

```python
# Compute Standard Error of Means and Variances
for ii in range(2):
    se_mean = np.sqrt(np.sum((means[ii] - mu_x)**2)/mciter)
    se_var  = np.sqrt(np.sum((variances[ii] - var_x)**2)/mciter)
    
    print("For %i %i-roll simulations: " % (mciter,n_rolls[ii]))
    print("\tStandard Error of the means is %f"    % se_mean)
    print("\tStandard Error of the variance is %f" % se_var)
```

## Q4 Sea Level and Barometric Pressure

```python
# Generate White Noise Timeseries
tmax   = 200
phi    = 0.60
std    = 1 #np.sqrt(0.01125)#0.0025
mciter = 100000

def make_ar1(phi,tmax,std,mciter):
    wn = np.random.normal(0,std,(mciter,tmax)) # [iteration, time]
    rn = np.zeros(wn.shape)
    for t in np.arange(1,tmax):
        rn[:,t] = rn[:,t-1]*phi + wn[:,t]
    return wn,rn

# Generate White Noise
wn_slp,rn_slp = make_ar1(phi,tmax,std,mciter)
wn_ssh,rn_ssh = make_ar1(phi,tmax,std,mciter)
#print(wn.shape),print(rn.shape)
```

```python
# Identify standard deviation thresholds (should theoretically be 1)
thres_slp = np.std(rn_slp-rn_slp.mean(1)[:,None])
thres_ssh = np.std(rn_ssh-rn_ssh.mean(1)[:,None])
print("1-stdev Threshold for SLP (SSH) is %f (%f)" % (thres_slp,thres_ssh))

# Identify higher than usual SSH and lower than usual SLP and counts
counts = np.zeros((mciter,3)) # 0 = counts of hi, 1 = counts of lo, 2 = both coincide
for mc in range(mciter):
    # Identify indices of events
    id_ssh_hi  = np.where(rn_ssh[mc,:] > thres_ssh)[0] # Higher than usual SSH
    id_slp_lo  = np.where(rn_slp[mc,:] < -thres_slp)[0] # Lower than usual SLP
    
    # Get Counts and record
    n_hi       = len(id_ssh_hi)
    n_lo       = len(id_slp_lo)
    n_coincide = len(np.intersect1d(id_ssh_hi,id_slp_lo))
    counts[mc,0] = n_lo
    counts[mc,1] = n_hi
    counts[mc,2] = n_coincide
```

```python
# Example Identification for 1 Iteration
mc = 22
fig,ax = plt.subplots(1,1,figsize=(16,4))

ax.plot(rn_slp[mc,:],color="r",label="SSH")
ax.plot(rn_ssh[mc,:],color="b",label="SST")

ax.axhline(thres_ssh,color="b",ls="dashed")
ax.axhline(-thres_slp,color="r",ls="dashed")

id_hi      = np.where(rn_ssh[mc,:] > thres_ssh)[0] # Higher than usual SSH
id_lo      = np.where(rn_slp[mc,:] < -thres_slp)[0] # Lower than usual SLP
n_coincide = len(np.intersect1d(id_hi,id_lo))

ax.scatter(id_hi,rn_ssh[mc,id_hi],marker="x",color="b")
ax.scatter(id_lo,rn_slp[mc,id_lo],marker="o",color="r")

for n in np.intersect1d(id_hi,id_lo):
    ax.axvline(n,c="gray",ls='dashed')
    
ax.legend()
ax.set_title("%i Coincident Events out of 200, P=%f" % (n_coincide,n_coincide/200))
ax.set_xlabel("Days")
```

```python
# Get 95% Confidence and plot results
def get_cf(wn,bnds=[0.025,0.975]):
    wn_srt = np.sort(wn)
    N      = wn.shape[0]
    ids    = [int(bnds[b]*N) for b in range(len(bnds))]
    cfs    = [wn_srt[ii] for ii in ids]
    return cfs

prob_coincide = counts[:,2]/200
cfs           = get_cf(prob_coincide)

fig,ax = plt.subplots(1,1)
ax.hist(prob_coincide,bins=20,edgecolor="w",alpha=0.5)
ax.axvline(cfs[0],label="2.5%"  + " = %f"%cfs[0],color="gray")
ax.axvline(cfs[1],label="97.5%" + " = %f"%cfs[1],color="gray")
ax.legend()

ax.axvline([0.055],label="Observed Value=0.055",color="red")

ax.set_title("Histogram (iterations=%i)" % mciter)
ax.set_xlabel("P(SSH+ and SLP-), n=%i" % (tmax))
ax.set_ylabel("Frequency")
```

## Q5. Damping Timescale and Sampling Rate

```python
txtfile  =  "/Users/gliu/Downloads/06_School/10_Fall2023/12S992/PSETs/PSET01/hw1data.txt"
data     = np.loadtxt(txtfile)
plt.plot(data),plt.title("Loaded Timeseries")
```

```python
%%time
# Generate White Noise Timeseries
mciter             = 10000
tmax               = 10000
std                = 1 #np.std(data)
wn_samples = np.zeros((mciter,tmax)) # [variable, sample, time]
for mc in range(mciter):
    wn_samples[mc,:] = np.random.normal(0,std,(tmax))
```

```python
# Compute Correlation Coefficients
wn_corr = np.zeros(mciter)
for mc in range(mciter):
    ts          = wn_samples[mc,:]
    wn_corr[mc] = np.corrcoef(ts[:-1],ts[1:])[0,1]
```

```python
# Test Script to get the confidence interval
# for a set record length

cfs = get_cf(wn_corr)

fig,ax = plt.subplots(1,1)
ax.hist(wn_corr,bins=25,edgecolor="w",alpha=0.5)
ax.axvline(cfs[0],label="2.5%"  + " Conf = %f"%cfs[0],color="gray")
ax.axvline(cfs[1],label="97.5%" + " Conf = %f"%cfs[1],color="gray")
ax.legend()

ax.set_title("Histogram (iterations=%i)" % mciter)
ax.set_xlabel("AR(1) Corr. Coeff. for White Noise Timeseries, n=%i" % (tmax))
ax.set_ylabel("Frequency")
```

```python
# Examine how the AR(1) coefficient decreases with subsample

subsample     = np.arange(1,101,1)
phi_sub       = np.zeros(subsample.shape)
record_length = np.zeros(subsample.shape)
for s,step in enumerate(subsample):
    ts_in            = data[::step]
    record_length[s] = len(ts_in)
    phi_sub[s]       = np.corrcoef(ts_in[:-1],ts_in[1:])[0,1]
```

```python
%%time
# For each sample size, perform Monte Carlo and get the 95% Interval

nlen      = len(record_length)
niter     = mciter
cfs_all   = np.zeros((nlen,2))
wn_coeffs = np.zeros((nlen,niter))
axcol     = "red"
for s in range(nlen):
    
    # Generate White Noise Timeseries
    tmax      = int(record_length[s])
    wn_ts_all = np.zeros((niter,tmax))
    for mc  in range(niter):
        # Generate White Noise Time Series
        ts              = np.random.normal(0,1,tmax)
        wn_ts_all[mc,:] = ts
        # Compute lag-1 coeff
        wn_coeffs[s,mc] = np.corrcoef(ts[:-1],ts[1:])[0,1]
    
    # Get Confidence Intervals
    coeffs = wn_coeffs[s,:] # iterations
    cfs    = get_cf(coeffs,bnds=[0.025,0.975])
    cfs_all[s,:] = cfs.copy()
```

```python
# Examine how the AR(1) coefficient decreases with subsample

# Compute the phi for each subsample
subsample     = np.arange(1,101,1)
phi_sub       = np.zeros(subsample.shape)
record_length = np.zeros(subsample.shape)
for s,step in enumerate(subsample):
    ts_in            = data[::step]
    record_length[s] = len(ts_in)
    phi_sub[s]       = np.corrcoef(ts_in[:-1],ts_in[1:])[0,1]


# Make the plot
fig,ax = plt.subplots(1,1)

# Plot the AR(1) coefficient and confidence interval by subsample length
l0=ax.plot(subsample,phi_sub,lw=0.75,label="AR(1) Corr. Coeff. (data)")
l1= ax.fill_between(subsample,cfs_all[:,0],cfs_all[:,1],alpha=0.26,color='gray',label="95% Confidence")
#ax.plot(subsample,cfs_all[:,1],color='gray',label="95% Confidence")
#ax.plot(subsample,cfs_all[:,0],color='gray',label="5% Confidence")
ax.set_xlabel("Subsampling Step")
ax.set_ylabel("AR(1) Correlation Coefficient ($\phi$)")
ax.grid(True,ls='dotted')
ax.set_title("Subsample Step vs. $\phi$")

# Plot the intersection to identify where it becomes indistinguishable
id_where = np.where(phi_sub < cfs_all[:,1])[0]
ax.axvline(subsample[id_where[0]],label="N=%i" % (subsample[id_where[0]]),color="grey",ls="dashed")

# Also plot the sample size
ax2 = ax.twinx()
ax2.plot(subsample,record_length,color=axcol,lw=0.7,ls='dashed',label="Sample Size")
l2 = ax2.set_ylabel("Sample Size",color=axcol)
ax2.spines['right'].set_color(axcol)
ax2.tick_params(axis='y', colors=axcol)

ax.legend()
#ax.axhline(cfs[1])
```
