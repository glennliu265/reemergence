```python
# Here's a function 

ts        = thickness.copy() #np.arange(0,100)
l         = 8

def median_smoother(ts,l,debug=False):
    tmax           = len(ts)
    ts_smooth      =  np.zeros(tmax) * np.nan
    window_lengths =  ts_smooth.copy()
    for t in range(tmax):
        istart = t-l # startpoint
        iend   = t+l # endpoint
        if t-l < 0: # Window crosses start point, set to zero
            istart = 0
        if t+l > (tmax): # Window crossing end point, set to max
            iend   = tmax
        sel_idx           = np.hstack([np.arange(istart,t),np.arange(t+1,iend)]) # Get Indices, excluding t
        sel_ts            = ts[sel_idx] # Select Times
        window_lengths[t] = len(sel_idx)
        ts_smooth[t]      = np.median(sel_ts) # Take the median of the window
        if debug:
            print("t=%i"%t)
            print(sel_idx)
    return ts_smooth,window_lengths
    
        
        
    
```

```python
half_lengths = np.arange(1,20)

smoothed_timeseries = []
window_lengths      = []
for l in range(len(half_lengths)):
    l = half_lengths[l]
    ts_smooth,winlen = median_smoother(thickness,l)
    smoothed_timeseries.append(ts_smooth)
    window_lengths.append(winlen)
smoothed_timeseries = np.array(smoothed_timeseries)
smoothed_timeseries.shape
```

```python
fig,ax = plt.subplots(1,1)
for l in range(len(half_lengths)):
    ax.plot(years,smoothed_timeseries[l,:],label="l=%i"%half_lengths[l])
ax.legend()
```
