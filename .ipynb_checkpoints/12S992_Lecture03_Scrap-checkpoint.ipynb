```python

def get_thres(start,window,timeseries):
    tper_idx = (time >= start) * (time <= start+window)
    sl_period = timeseries[tper_idx]
    thres  = np.nanpercentile(sl_period,99)
    return thres
    


def count_events(thres,return_all=False):
    tper0 = (time >= 1960) * (time <= 1980)
    tper1 = (time >= 1995) * (time <= 2014)
    
    count0=len(np.where(vals[tper0] > thres)[0])
    count1=len(np.where(vals[tper1] > thres)[0])
    if return_all:
        return count1,count0,count1/count0
    return count1/count0

    

```

```python
starts = np.arange(1960,1996)
ends   = np.arange(1980,2015)
window = 30
thres_all    = []
ratios       = []
count0 = []
count1 = []

for ii in range(len(starts)):
    thres = get_thres(starts[ii],window,vals)
    thres_all.append(thres)
    print(thres)
    cnt1,cnt0,ratio = count_events(thres,return_all=True)
    count0.append(cnt0)
    count1.append(cnt1)
    ratios.append(ratio)
    
    
```

```python
axcol='red'
fig,ax = plt.subplots(1,1,constrained_layout=True,figsize=(12,4))

ax.plot(starts,ratios)
ax.set_title("Ratio of Extreme SSH Events (>99 percentile)")
ax.set_xlabel("Starting Year of %i-year window" % window)
ax.set_ylabel("Ratio (1995-2014 / 1960-1980)")

ax2 = ax.twinx()
ax2.plot(starts,thres_all,color=axcol)
l2 = ax2.set_ylabel("99th Percentile Threshold",color=axcol)
ax2.spines['right'].set_color(axcol)
ax2.tick_params(axis='y', colors=axcol)

ax.grid(True)
```
